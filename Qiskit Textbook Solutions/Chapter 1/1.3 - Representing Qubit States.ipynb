```python
%matplotlib inline
# Importing standard Qiskit libraries
from qiskit import QuantumCircuit, execute, Aer, IBMQ
from qiskit.compiler import transpile, assemble
from qiskit.visualization import *
from math import sqrt, pi

# Loading your IBM Q account(s)
provider = IBMQ.load_account()
```

## Quick Exercise 1 

### Defining the Circuit

```python
qc = QuantumCircuit(1,1)
qc.draw('mpl')
```

### 1. Create a state vector that will give a 1/3 probability of measuring |0‚ü©.

```python
initialState = [1/sqrt(3), sqrt(2/3)]
qc.initialize(initialState, 0)
qc.measure(0, 0)
qc.draw('mpl')
```

### Simulating the Circuit

```python
nativeSim = Aer.get_backend('qasm_simulator')
result = execute(qc, backend = nativeSim, shots = 1024).result()
plot_histogram(result.get_counts(qc))
```

### 2. Create a different state vector that will give the same measurement probabilities.

```python
qc = QuantumCircuit(1, 1) 
alternateState = [sqrt(3) / 3, 2 /sqrt(6)]
qc.initialize(alternateState, 0)
qc.measure(0, 0)
qc.draw('mpl')
```

```python
result = execute(qc, backend = nativeSim, shots = 1024).result()
plot_histogram(result.get_counts(qc))
```
